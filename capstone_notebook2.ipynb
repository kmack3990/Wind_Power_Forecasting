### Feature Engineering / Machine Learning / FBProphet
In Notebook 2, we will add a new feature called  `four_turbine_avg_power` that is the average power output of all four turbines in EDP’s Wind Farm 1. We will also transform our data set to create our independent variable as a series of time steps prior to our target to make a window that can move over our dataframe and make one step predictions.  We will evaluate this process with the Random Forest Regressor. 

Lastly, we will see how a FBProphet model performs with our average power signal.  The FBProphet model was developed to model nonlinear times series and could prove useful in our purpose.

```python
# Adopted from Scikit-Learn Tutorial on Time Series Analysis
# https://scikit-learn.org/stable/auto_examples/applications/plot_cyclical_feature_engineering.html#time-related-feature-engineering
# We could not get the median_absolute_error to work on the cross val. It is our least insightful 
# metric and we feel comfortable leaving it out. 
def evaluate(model, X, y, cv):
    '''
    Prints out a R-squared and MAE for run on a supervised learning model with a 
    cross-validation split. 

    Parameters
    ----------
    model: model used 
    x: independent variable as a data timeseries dataframe
    y: dependent variable as a data timeseries dataframe

    Returns
    -------
    Print statement of R-squared mean and standard deviation and mean absolute error mean and 
    standard deviation for the test set of each test train split. 

    Examples
    --------
    >>>> RFRmodel, X, y, cv=ts_cv_2
    >>>> R-Squared:    0.781 +/- 0.260129
         Mean Absolute Error:     111.439 +/- 70.953
    '''
    
    cv_results = cross_validate(
        model,
        X,
        y, 
        cv=cv,
        scoring=["r2","neg_mean_absolute_error"],
    )
    mae = -cv_results["test_neg_mean_absolute_error"]
    r2 = cv_results["test_r2"]
    print(
        f"R-Squared:    {r2.mean():.3f} +/- {r2.std():3f}\n"
        f"Mean Absolute Error:     {mae.mean():.3f} +/- {mae.std():.3f}\n"
    )
```

# Random Forest Regressor with Time Lags - Daily Data
- The ideas for this RFR were inspired by two articles written by Jason Brownlee, PhD: 

Title: Random Forest for Time Series Forecasting

https://machinelearningmastery.com/random-forest-for-time-series-forecasting/

Title: How to Convert a Time Series to a Supervised Learning Problem in Python

https://machinelearningmastery.com/convert-time-series-supervised-learning-problem-python/

- We will just use `four_turbine_avg_power` as our X and y but make successive time lags one day at a at time to create our supervised learning model.  
- We employ Brownlee's function series_to_supervised to reshape our data frame so that each power data point will have "n_in" or "n_steps" number of previous data points or lags in the times series within the same row or series within an array.  This will create our independent variable, X, as the previous data points in the series.  The dependent variable, y, will be the original data point from which the time steps or lags were developed. An illustration of a table made by this function is in the cell below. Where `var1(t)` would be our dependent variable and the two timesteps back, `var1(t-1)` and `var1(t-2)`, would be the independent variable for that data point.  This process will create rows or series at the beginning of the data set containing nulls for each time step back that you make.  These rows of nulls will then be dropped from the data set.   

```python
# transform a time series dataf frame into a supervised learning array
    #series_to_supervised3 returns array 
# labeled series_to_supervised3 as we used two previous iterations before settling on this function.
def series_to_supervised3(data, n_in=1, n_out=1, dropnan=True):
    '''
    Frames a time series as a supervised learning dataset.
    Parameters
    ----------
        data: Dataframe
        n_in: Number of lag observations as input (X).
        n_out: Number of observations as output (y).
        dropnan: Boolean whether or not to drop rows with NaN values.
    Returns
    -------
        Array of series framed for supervised learning.
    Examples
    --------
    >>>> dataframe of one timeseries signal, n_in = 3, n_out = 1, dropnan = True
    >>>> Array with series for each n_out, rows of nulls will be dropped
    '''
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols = list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
    # put it all together
    agg = pd.concat(cols, axis=1)
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
    return agg.values
```

```python
# split a univariate dataset into train/test sets
def train_test_split(data, n_test):
    '''
    Splits data into a train and test given how many data points you want in the test. 
    Parameters
    ----------
        data: Array of supervised learning data created by the series_to_supervised3 function
        n_test: number of independent variables in your test set that will enter 
                walk_forward_validation
    Returns
    -------
        array for the train set and an array for the test set
    '''
    return data[:-n_test, :], data[-n_test:, :]
```

```python
def random_forest_forecast(train, testX):
     '''
    Fits an random forest model and makes a one step prediction value. 
    Parameters
    ----------
        train: list of training data
        testX: this is the series that is created in walk_forward_validation
    Returns
    -------
        yhat: a one step prediction value
    '''
    # transform list into array
    train = np.asarray(train)
    # split into input and output columns
    trainX, trainy = train[:, :-1], train[:, -1]
    # fit model
    model = RandomForestRegressor(n_estimators=1000,  random_state=32)
    model.fit(trainX, trainy)
    # make a one-step prediction
    yhat = model.predict([testX])
    return yhat[0]
```

```python
# walk-forward validation for univariate data
def walk_forward_validation(data, n_test):
    '''
    First, this function creates a train test split using the train_test_split function.  
    Then creates list of from the array of training data as history.  Then makes a list of 
    numbered "dates" for each value in the test set.  The function then loops through each
    series in the test set creating a test X series and test y value. It then utilizes the
    random_forest_forecast function to train a random forest regression model and also make
    a one step prediction.  The test series is then appended to the history to through the 
    for loop again to make another one step prediction for each series in test set.  
    Each prediction is appended to a predictions list.  Each predicted and expected 
    value is printed within the test set during modeling. After iterating through the for loop, 
    the function utilizes a variety of error metrics to analyze the final predictions list
    against the test series. The function then returns each metric, test sets with out their 
    independent variables, list of predictions and the numbered date in the test set. 
    Parameters
    ----------
        data: array - returned from series_to_supervised function
        n_test: number of data points, dependent variable time steps, to include in your test 
        set at the end of the end of the input array as the 'data' parameter
    Returns
    -------
         error: mean absolute error 
         test[:, -1]: test set with out each dependent variable
         predictions: list of predicted values for the length of the test set
         r_square: R-squared metric 
         med_ae: median absolute error
    '''

    predictions = list()
    # split dataset
    train, test = train_test_split(data, n_test)
    # seed history with training dataset
    history = [x for x in train]
    test_df = pd.DataFrame(test)
    #Got date number but could not get date still nice to have number 
    date=[date for date in test_df.index]
    # step over each time-step in the test set
    for i in range(len(test)):
        # split test row into input and output columns
        testX, testy = test[i, :-1], test[i, -1]
        # fit model on history and make a prediction
        yhat = random_forest_forecast(history, testX)
        # store forecast in list of predictions
        predictions.append(yhat)
        # add actual observation to history for the next loop
        history.append(test[i])
        # summarize progress 
        print('>expected=%.1f, predicted=%.1f' % (testy, yhat))
        
    # estimate prediction error
    error = mean_absolute_error(test[:, -1], predictions)
    r_square  = r2_score(test[:, -1], predictions)
    med_ae = median_absolute_error(test[:, -1], predictions)
    return error, test[:, -1], predictions, r_square, med_ae, date
```
